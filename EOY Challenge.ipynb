```python
##Importing Matplotlib to edit and create graph.
import matplotlib.pyplot as plt
```

```python

petrolStats = [442583,351174]
petrolNames=["Petrol 2022","Petrol 2021"]

dieselStats = [93087,46028]
dieselNames=["Diesel 2021","Diesel 2022"]

electricStats = [73893,115249]
electricNames = ["Electric 2021","Electric 2022"]
```

```python
plt.figure(figsize=(13,13))

plt.style.use('fast')



petrolBar=plt.bar(petrolNames,petrolStats, label="Petrol Vehicles",color='purple',width=0.5)

plt.bar_label(petrolBar,labels=petrolStats, label_type="edge", padding = 5, fontsize=12)



dieselBar=plt.bar(dieselNames,dieselStats,label="Diesel Vehicles",width=0.5)

plt.bar_label(dieselBar,labels=dieselStats, label_type="edge", padding = 5, fontsize=12)


electricBar =plt.bar(electricNames,electricStats, label="Electric Vehicles", color="green",width=0.5)
plt.bar_label(electricBar,labels=electricStats, label_type="edge", padding = 5,fontsize=12)

plt.title("Impact of ULEZ Expansion on Vehicles", pad= 13, fontsize=15)


plt.xlabel("Type Of Fuel", labelpad=20, fontsize=15)
plt.ylabel("Number Of Cars Registered", labelpad=25, fontsize=15)
plt.yticks([0,50000,100000,150000,200000,250000,300000,350000,400000,450000], fontsize=14)
plt.xticks(fontsize=14)
plt.legend()
plt.show()



```
