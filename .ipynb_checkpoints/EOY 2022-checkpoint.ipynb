```python
##Importing standard functions used to read the csv and create the bar graph


import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Attaching a variable to make it easily accessible

figures=pd.read_csv("figures.csv")
```

```python
##SETTING WHAT INDEX THE TABLE STARTS FROM

figures.set_index("FUEL TYPE", inplace=True)
```

```python
##DISPLAYING TABLE
figures
```

```python

##THEME
plt.style.use('fast')

##CREATING BAR CHART
figures.plot(kind="bar", color="blue", fontsize=14, figsize=(10,8))

##EDITING LEGEND
plt.legend(fontsize=12)

##EDITING X LABEL
plt.xticks(rotation = 360, fontsize=12)

##ADDING TITLE
plt.title("Impact of ULEZ Expansion On Vehicles", pad= 50, fontsize=18,fontstyle='oblique')

#DEFINING Y & X LABELS
plt.ylabel("Yearly Registrations of Vehicles",fontsize=15,labelpad=30)
plt.xlabel("Type of Fuel",fontsize=15,labelpad=30)


```
