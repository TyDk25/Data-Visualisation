```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
figures=pd.read_csv("figures.csv")
```

```python
figures.set_index("FUEL TYPE", inplace=True)
```

```python
figures
```

```python
plt.style.use('fast')

figures.plot(kind="bar", color="blue", fontsize=14, figsize=(10,8))

plt.legend(fontsize=12)

plt.xticks(rotation = 360, fontsize=12)

plt.title("Impact of ULEZ Expansion On Veichles", pad= 50, fontsize=18,fontstyle='oblique')

plt.ylabel("Number of cars registered",fontsize=15,labelpad=30)

plt.xlabel("Type of Fuel",fontsize=15,labelpad=30)
plt.figure(figsize=((300,20)))
```
