```python
##Importing standard functions used to read the csv and create the bar graph
import pandas as pd
import matplotlib.pyplot as plt
```

```python
petrolStats = [442583,351174]
petrolNames=["Petrol 2022","Petrol 2021"]
dieselStats = [93087,46028]
dieselNames=["Diesel 2021","Diesel 2022"]
electricStats = [73893,115249]
electricNames = ["Electric 2021","Electric 2022"]
```

```python
plt.figure(figsize=(10,10))
plt.style.use('fast')
plt.bar(petrolNames,petrolStats, label="Petrol Figures",color='purple',width=0.5)
plt.bar(dieselNames,dieselStats,label="Diesel Figures",width=0.5)
plt.bar(electricNames,electricStats, label="Electric Vehicles", color="green",width=0.5)

plt.title("Impact of ULEZ Expansion on Vehicles", pad= 13)
plt.xlabel("Type Of Fuel", labelpad=20)
plt.ylabel("Number Of Cars Registered", labelpad=25)
plt.legend()
plt.show()



```
